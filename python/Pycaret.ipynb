```python
#%pip install pycaret
```

```python
from pycaret.classification import *
classification = setup(
    session_id=6698,
    data=df,
    target='label',
    train_size=0.7,
    preprocess=False,
    use_gpu=False,
)
```

```python
top3 = compare_models(n_select=3)
```

```python
print(top3)
```

```python
melhores_modelos = top3[:3]
```

```python
modelo1 = melhores_modelos[0]
```

```python
print(modelo1)
```

```python
evaluate_model(modelo1)
```
