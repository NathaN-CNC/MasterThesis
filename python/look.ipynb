```python
import pandas as pd
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
def count_changes(df):
  counts = {}
  prev = {}

  def _count_changes(row):
    key = row["name"] + "," + row["type"]
    if key not in counts:
      counts[key] = 0
      prev[key] = row["skill"]
    if prev[key] != row["skill"]:
      counts[key] += 1
      prev[key] = row["skill"]
    return counts[key]

  df["changes"] = df.apply(_count_changes, axis=1)
```

```python
metrics = pd.read_csv(r"C:\Users\natha\Documents\AgentesML\jade_experiments\env\metrics\0.07\metrics_sample_c075_a200_api0.07.csv")
metrics["datetime"] = metrics["datetime"].str.slice(0, 22).str.slice_replace(11, 14, '')
metrics["datetime"] = pd.to_datetime(metrics['datetime'], format="%Y-%m-%dT%H:%M:%S") # 2024-04-21T17:17:44:10.202
metrics["name_skill"] = metrics.apply(lambda row: f'{row["name"]} ({row["skill"]})', axis=1)
count_changes(metrics)
metrics.describe()
```
