```python
import pandas as pd
```

```python
dtest = pd.read_csv('C:\\EstudosML\\ML distribuído\\UNSW-N15\\Dataset\\UNSW_NB15_training-set.csv')

dtest = dtest.drop(columns=['id'])
dtest = dtest.drop(columns=['attack_cat'])

#A coluna service possui valores '-', vamos trocar para None
dtest['service'] = dtest['service'].apply(lambda x:"None" if x=="-" else x)

#como temos as features proto, service e state como categóricas, temos que transformálas em binários
#one hot enconder
dtest_final = pd.get_dummies(dtest, columns = ['proto', 'service', 'state'])
```

```python
dtest_final.to_csv('dadosteste.csv', index = False)
```
